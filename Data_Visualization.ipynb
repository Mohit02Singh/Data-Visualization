Adding Labels

Labels are annotations or descriptions: There many types of labels:
* Title, Axis, Data labels
* Legend , Tick(Nemric or categoric labels)
* Color bar or Figure labels


```python
#Adding Labels  to the plot
import matplotlib.pyplot as plt

days = range(1, 11)
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1, 35.2, 40.5]

fig, ax = plt.subplots()
#Add marker
ax.plot(days, celsius_values, marker='o', linestyle='-')
ax.plot(days, celsius_values)

ax.set(xlabel='Day',
       ylabel='Temperature in Celsius',
       title='Temperature Graph')
#Adding Specific color for any label,title etc.
ax.xaxis.label.set_color('red') # Axis labels with color
ax.yaxis.label.set_color('blue')
ax.title.set_color('green')# Title label with color
ax.grid(True, color='red')# Grid with color
plt.show()
# It's attractive after give labels or colors or easily understand ,percieve as a "Layman"
```

```python
#Scatter plot
import matplotlib.pyplot as plt

days = list(range(1, 9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)#Subplots with the help of using syntax subplots

# Plot for minimum temperature
ax1.set_ylabel("Min Temperature (Celsius)")
ax1.plot(days, celsius_min, "oy", label="Min Temperature")#We can give labels color to effective visualize as a Layman , Syntax= color='r' like
ax1.grid(True,color=('Red'))
ax1.legend()

# Plot for maximum temperature
ax2.set_xlabel("Day")
ax2.set_ylabel("Max Temperature (Celsius)")
ax2.plot(days, celsius_max, "or", label="Max Temperature")
ax2.grid(True,color=('Red'))#Grid color , we used other colors(red,green,blue,orange etc.)
ax2.legend()

plt.suptitle("Temperature Graphs")  # Add a title for the entire figure

plt.show()
```

```python
#Radar chart
import matplotlib.pyplot as plt
import numpy as np

# Data for stack plot
idxes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [23, 42, 33, 43, 8, 44, 43, 18, 21]
y2 = [9, 31, 25, 14, 17, 17, 42, 22, 28]
y3 = [18, 29, 19, 22, 18, 16, 13, 32, 21]

# Create a polar plot for the radar chart
theta = np.linspace(0, 2*np.pi, len(idxes), endpoint=False)
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

# Plot each dataset on the radar chart
ax.plot(theta, y1, label='Dataset 1')
ax.fill(theta, y1, alpha=0.25)

ax.plot(theta, y2, label='Dataset 2')
ax.fill(theta, y2, alpha=0.25)

ax.plot(theta, y3, label='Dataset 3')
ax.fill(theta, y3, alpha=0.25)

# Set the labels and title
ax.set_xticks(theta)
ax.set_xticklabels(idxes)# We also give labels colors
plt.title('Radar Chart Example')

# Display the legend
plt.legend()

plt.show()
```

```python
# Bubble chart
import matplotlib.pyplot as plt

# Data for bubble chart
idxes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [23, 42, 33, 43, 8, 44, 43, 18, 21]
y2 = [9, 31, 25, 14, 17, 17, 42, 22, 28]
y3 = [18, 29, 19, 22, 18, 16, 13, 32, 21]
bubble_sizes = [30, 50, 20, 40, 10, 60, 45, 25, 35]  # Third variable for bubble size

# Create a bubble chart
plt.scatter(idxes, y1, s=bubble_sizes, label='Dataset 1')
plt.scatter(idxes, y2, s=bubble_sizes, label='Dataset 2')
plt.scatter(idxes, y3, s=bubble_sizes, label='Dataset 3')

# Set the labels and title
plt.title('Bubble Chart Example')
plt.xlabel('Index Values')
plt.ylabel('Y values')

# Display the legend
plt.legend()

plt.show()
```

```python
# Treemap
import matplotlib.pyplot as plt
import squarify# Is used to create a treemap and plot

# Data for treemap
idxes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [23, 42, 33, 43, 8, 44, 43, 18, 21]
y2 = [9, 31, 25, 14, 17, 17, 42, 22, 28]
y3 = [18, 29, 19, 22, 18, 16, 13, 32, 21]

# Calculate total values for each dataset
total_y1 = sum(y1)
total_y2 = sum(y2)
total_y3 = sum(y3)

# Create labels for each dataset
labels = ['Dataset 1\n{:.1%}'.format(value/total_y1) for value in y1]
labels += ['Dataset 2\n{:.1%}'.format(value/total_y2) for value in y2]
labels += ['Dataset 3\n{:.1%}'.format(value/total_y3) for value in y3]

# Create a treemap
squarify.plot(sizes=y1 + y2 + y3, label=labels, color=['red', 'green', 'blue'], alpha=0.7)# Give colors to labels to effectiveness
#Size- Give rectangle size
#Labels- percentage contibution
#alpha- Adjust the transparency
# Remove axis labels
plt.axis('off')

# Set the title
plt.title('Treemap Example')

plt.show()
```

Pie Charts

Pie charts used:
* It represents the specific category or component and all pie represent the total of a component
* Percentage
* Proportional distribution
* Market share
* Project Task distribution
* Demographic composition
* Contribution in revenue

```python
#Pop out pie slice
import matplotlib.pyplot as plt

labels = 'C', 'Python', 'Java', 'C++', 'C#'
sizes = [13.38, 11.87, 11.74, 7.81, 4.41]

# Define explode to specify the fraction of the radius with which to offset each wedge
explode = (0.1, 0, 0, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=0)
ax1.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle

plt.title('Popularity of Different Programming Languages')
plt.show()
```

```python
#Without pop out pie slice
import matplotlib.pyplot as plt

labels = 'C', 'Python', 'Java', 'C++', 'C#'
sizes = [13.38, 11.87, 11.74, 7.81, 4.41]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=0)
ax1.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle

plt.title('Popularity of Different Programming Languages')
plt.show()
```

```python
# Doughnut Chart
import matplotlib.pyplot as plt

# Data for both pie and doughnut charts
labels = 'C', 'Python', 'Java', 'C++', 'C#'
sizes = [13.38, 11.87, 11.74, 7.81, 4.41]

# Create a pie chart
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle
plt.title('Popularity of Different Programming Languages (Pie Chart)')
plt.show()

# Create a doughnut chart
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))# Width means create the hole of doughnut
ax2.axis('equal')  # Equal aspect ratio ensures that the doughnut is drawn as a circle
plt.title('Popularity of Different Programming Languages (Doughnut Chart)')
plt.show()
```

In the doughnut chart is good but not more attractive like pie chart according to me.

```python
#Stacked bar chart
import matplotlib.pyplot as plt

# Data for stacked bar chart
labels = ['C', 'Python', 'Java', 'C++', 'C#']
sizes = [[13.38], [11.87], [11.74], [7.81], [4.41]]  # Each sublist represents a category

# Create a stacked bar chart
fig, ax = plt.subplots()
ax.bar(labels, sizes[0], label='C')
bottoms = sizes[0]  # Keep track of the bottom values for stacking

for i in range(1, len(sizes)):
    ax.bar(labels, sizes[i], bottom=bottoms, label=labels[i])
    bottoms = [sum(x) for x in zip(bottoms, sizes[i])]  # Update the bottom values for stacking

plt.title('Popularity of Different Programming Languages (Stacked Bar Chart)')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.legend()
plt.show()
```

Satcked bar chart is better than pie chart according to me . As a layman it's easy to understand and visualize the data .

```python
# Stacked area chart
import matplotlib.pyplot as plt

# Data for stacked area chart
labels = ['C', 'Python', 'Java', 'C++', 'C#']
sizes = [[13.38], [11.87], [11.74], [7.81], [4.41]]  # Each sublist represents a category

# Transpose the data for stackplot
stack_data = list(zip(*sizes))

# Custom colors for each category
colors = ['red', 'green', 'blue', 'orange', 'purple']

# Create a stacked area chart
fig, ax = plt.subplots()
ax.stackplot(labels, *stack_data, labels=labels,colors=colors)#stackplot used to plot and we can give color according to choice. What type of color is good for visualize?

plt.title('Popularity of Different Programming Languages (Stacked Area Chart)',color='r')
plt.xlabel('Programming Languages')#Labels and also give color
plt.ylabel('Popularity (%)')
plt.legend(loc='upper left')# We can shift the legend according to choice
plt.show()
```

```python
#Horizontal bar charts
import matplotlib.pyplot as plt

# Sample data
languages = ['C', 'Python', 'Java', 'C++', 'C#']
popularity = [13.38, 11.87, 11.74, 7.81, 4.41]

#Give Colors for effective visualize
colors=['skyblue','red','yellow','green','purple']

# Create a horizontal bar chart
plt.barh(languages, popularity, color=colors)#We can give color in different bar for different color

# Add labels and title
plt.xlabel('Popularity (%)')# Labels colors also give by a user to better visualize
plt.ylabel('Programming Languages')
plt.title('Popularity of Different Programming Languages')

# Show the plot
plt.show()
```

It is more attractive than pie chart and easily understand as layman.
